#### Entfernen redundanter Spalten:
* Die Spalte `title`, die dieselben Informationen wie die Spalte `channel_name` enthält, wird entfernt. Dies vermeidet Redundanz und minimiert die Datenmenge, zum Verwalten und Analysieren. 
* Die aktuelle `rank`-Spalte, die den Rang des YouTube-Kanals basierend auf der Anzahl der Abonnenten angibt, ist durch das vorherige Löschen von Zeilen nicht mehr konsistent. Um eine präzise Rangordnung zu gewährleisten, werde ich die aktuelle `rank-`Spalte entfernen, den DataFrame nach den Abonnentenzahlen neu sortieren und den Index, sowie `rank` daraufhin neu setzen.

```python
# Entfernen der aktuellen rank-Spalte
df.drop(columns=['rank'], inplace=True)
df.drop(columns=['title'], inplace=True)

# Neuordnung des DataFrames
df.sort_values(by='subscribers', ascending=False, inplace=True)

# Neusetzen des Indexes und Rang
df.reset_index(drop=True, inplace=True)
df['rank'] = df.index + 1
```
