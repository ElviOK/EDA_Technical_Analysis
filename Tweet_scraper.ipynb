```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Connect to Website 
URL = 'https://api.stocktwits.com/widgets/stream?width=300&amp;height=1700&amp;domain=finviz.com&amp;scrollbars=0&amp;streaming=true&amp;header=0&amp;limit=24&amp;symbol=DB&amp;border_color=22262F&amp;border_color_2=22262F&amp;box_color=22262F&amp;header_text_color=C3C6D0&amp;divider_color=353945&amp;stream_color=22262F&amp;username_color=F3F3F5&amp;username_hover_color=F3F3F5&amp;username_font=Verdana%2C%20Arial%2C%20Tahoma&amp;username_size=12&amp;divider_type=solid&amp;link_color=57aefb&amp;link_hover_color=57aefb&amp;font=Verdana%2C%20Arial%2C%20Tahoma&amp;font_size=12&amp;text_color=C3C6D0&amp;time_color=868EA5&amp;time_font_size=10'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36 OPR/100.0.0.0","Accept-Encoding": "gzip, deflate, br","Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7","DNT":"1","Connection":"close","Upgrade-Insecure-Requests":"1"}

# Pulling in the content
page = requests.get(URL, headers=headers)

soup = BeautifulSoup(page.content, "html.parser")

columns = soup.find_all(['span','a'], class_=['time','body','username']) 
```

```python
values_table = [column.text.strip() for column in columns]
print(values_table)
```

```python
columns_table = ['username','time','body']
df = pd.DataFrame(columns = columns_table)

df
```

```python
start = 0 
end = len(values_table)
step = 3
for i in range(start, end, step):
    x = i
    individual_row_data = values_table[x:x+step]
    #print(individual_row_data)
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
#Adding year to df
df["time"] = "2023 " + df["time"]

#Converting Object Date into Pandas DateFrame format
df["time"] = pd.to_datetime(df["time"], format="%Y %b. %d at %I:%M %p")
```
